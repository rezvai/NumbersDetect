# 1.1 Предобработка обучающих данных

```python
# Импортируем библиотеку для работы с файловой структурой
import os

# Создаем списочный генератор, где будут храниться все иображения
images = [os.path.join('путь до директории', x) for x in os.listdir('путь до директории') if x.endswith('тип файла')]
# Создаем списочный генератор, где буду храниться все аннотации
labels = [os.path.join('путь до директории', x) for x in os.listdir('путь до директории') if x.endswith('тип файла')]

# Сортиурем созданные списки
images.sort()
labels.sort()
```

```python
# Из библиотеки с готовыми алгоритмамим под машинное обучение импортируем модуль ля рабиение выборки
from sklearn.model_selection import train_test_split
# Импортиурем бибилиотеку для работы с массивам
import numpy as np

# Разбиваем нашу обучающую выборку на train и val
train_images, val_images, train_labels, val_labels = train_test_split(images, labels, test_size=0.2, random_state=42)
# Разбиваем выборку val пополам на test и val
val_images, test_images, val_labels, test_labels = train_test_split(val_images, val_labels, test_size=0.5, random_state=42)

# Выводим информацию о получившихся выборках train, val, test:
print(f'Количество элементов в train выборке: {np.array(train_images).size} \
      \nВсе элементы train выборки: {sorted(train_images + train_labels)}', end='\n\n')
print(f'Количество элементов в val выборке: {np.array(val_images).size} \
      \nВсе элементы val выборки: {sorted(val_images + val_labels)}', end='\n\n')
print(f'Количество элементов в test выборке: {np.array(test_images).size} \
      \nВсе элементы v выборки: {sorted(test_images + test_labels)}')
```

```python
# Создаем переменную, в котрой будет храниться название папки нащего датасета
dataset= 'dataset'
# Созаем кортеж, в котором будем хранить названия всех путей нашего датасета
dataset_dirs = ('images/train', 'images/val', 'images/test', 'labels/train', 'labels/val', 'labels/test')
# Создаем цикл, проверяющий наличие данных папок, при их отсуствие он их создает
for dir in dataset_dirs:
    # Создаем путь до папки
    path = os.path.join(dataset, dir)
    # Если такого пути нет, то создаем его
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
# Импортируем библиотеку для копирования файлов
import shutil

# Создаем функцию для копирования файлов из одной директории в другую
def move_files_to_folder(list_of_files, destination_folder):
    # Для всех файлов из переменной
    for file in list_of_files:
        # Пытаемся копировать, если происходит ошибка, то выводится в консоль: "Ошибка копирования: имя_файла.расширение"
        try:
            shutil.copy(file, destination_folder)
        except:
            print(f'Ошибка копирования: {file}')

# Создаем многомерный массив со всеми файлами внтури
all_the_files = [[train_images, val_images, test_images], [train_labels, val_labels, test_labels]]
# Создаем кортеж совсеми переменными нашей выборки
variabels_dataset = ('images', 'labels')
# Создаем массив со всеми задачами 
tasks = ['train', 'val', 'test']

# Создаем цикл, использующий написанную ранее функцию для копирования файлов
for i in range(len(variabels_dataset)):
    path = os.path.join(dataset, variabels_dataset[i])
    for j in range(len(tasks)):
        move_files_to_folder(all_the_files[i][j], os.path.join(path, tasks[j]))
```
